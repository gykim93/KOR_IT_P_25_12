```python
import pandas as pd
import numpy as np

```

```python
# 실습 데이터 생성
data = {
    '이름' : ['철수', '영희', '민수', '지영', '동수'],
    '학년' : [1, 2, 1, 3, 2],
    '성별' : ['남', '여', '남', '여', '남'],
    '국어' : [90, 80, 85, 95, np.nan], # np.nan => 결측치(데이터 없음)
    '영어' : [85, 90, 80, 95, 70],
    '수학' : [100, 88, 92, 85, 60]
}
df = pd.DataFrame(data)
df.to_csv('students.csv', index=False)
# print("실습 데이터 생성!")
```

```python
# 결측치 처리 : 비어있는 점수(Nan)을 0점으로 채우기
df['국어'] = df['국어'].fillna(0)

# 총점 컬럼 새로 만들기
df['총점'] = df['국어'] + df['영어'] + df['수학']

# 그룹화 : 학년별로 국어 영어 수학 평균 구하기
grade_mean = df.groupby('학년')[['국어', '영어', '수학']].mean()

# 학년별 평균 점수
# display(grade_mean)

# 위에 분석한 결과를 새 파일로 저장
df.to_csv('result.csv', index=False)
```
